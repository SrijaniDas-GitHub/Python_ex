```python
class Card:
    def __init__(self, suit, face, value):
        self.suit = suit
        self.face = face
        self.value = value
        
    
class Deck:
    suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
    #suits = ('Hearts', 'Diamods')
    cards = [['Two', 2], ['Three', 3], ['Four', 4], ['Five', 5], ['Six', 6], ['Seven', 7], ['Eight', 8], ['Nine',9], ['Ten', 10], ['Jack', 11], ['Queen', 11], ['King', 12], ['Ace', 13]]
    all_cards = []
    for suit in suits:
        for card in cards:
            all_cards.append(Card(suit, card[0], card[1]))

deck = Deck()
```

```python
class Player:
    def __init__(self, self_cards):
        self.self_cards = self_cards
        self.face_on = 0
        
    def wins_cards(self, cards):
        if type(cards) == type([]):
            self.self_cards.extend(cards)
        else:
            self.self_cards.append(cards)
        
    def draws_card(self):
        self.cards_at_stake = []
        print("Not at WAR!")
        self.cards_at_stake = [self.self_cards[0]]
        self.face_on = self.self_cards.pop(0)
```

```python
player_one_cards = deck.all_cards[:26]
player_two_cards = deck.all_cards[26:]
```

```python
player_one = Player(player_one_cards)
player_two = Player(player_two_cards)
```

```python
class Start_Game:
    def __init__(self, playerA, playerB):
        player_one = playerA
        player_two = playerB
        
    def start(self):
        i = 0
        while True:
            i += 1
            
            if len(player_one.self_cards) == 0:
                print("Player two wins")
                break
            if len(player_two.self_cards) == 0:
                print("Player one wins")
                break
                
            if i > 500:
                print("Round 500")
                break         
            
            player_one.draws_card()
            player_two.draws_card()  

            print("Player one got value : ", player_one.face_on.value)
            print("Player two got value : ", player_two.face_on.value)

            cards_to_win = player_one.cards_at_stake + player_two.cards_at_stake

            print("Player one cards at stake -", len(player_one.cards_at_stake))
            print("Player two cards at stake -", len(player_two.cards_at_stake))

            cards_to_win = player_one.cards_at_stake + player_two.cards_at_stake
            print(len(cards_to_win))

            if player_one.face_on.value > player_two.face_on.value:
                player_one.wins_cards(cards_to_win)
                print("Round %d : Player one won %d cards " %(i, len(cards_to_win)))        
            elif player_one.face_on.value < player_two.face_on.value:
                player_two.wins_cards(cards_to_win)
                print("Round %d : Player two won %d cards " %(i, len(cards_to_win)))
            else:
                player_one.cards_at_stake=[player_one.face_on]
                player_two.cards_at_stake=[player_two.face_on]

                while True:
                    print("At war!")
                    player_one.cards_at_stake += player_one.self_cards[:4]
                    player_two.cards_at_stake += player_two.self_cards[:4]

                    print("Player one cards at stake -", len(player_one.cards_at_stake))
                    print("Player two cards at stake -", len(player_two.cards_at_stake))

                    player_one.self_cards = player_one.self_cards[4:]
                    player_two.self_cards = player_two.self_cards[4:]

                    #run = 1
                    #if run > 5:
                    #    break

                    if player_one.cards_at_stake[-1].value > player_two.cards_at_stake[-1].value:
                        player_one.self_cards.extend(player_one.cards_at_stake + player_two.cards_at_stake)
                        print("Player one wins this round!!")
                        print(len(player_one.self_cards))
                        print(len(player_two.self_cards))
                        break
                    elif player_one.cards_at_stake[-1].value < player_two.cards_at_stake[-1].value:
                        player_two.self_cards.extend(player_one.cards_at_stake + player_two.cards_at_stake)
                        print("Player two wins this round!!")
                        print(len(player_one.self_cards))
                        print(len(player_two.self_cards))
                        break
                    else:
                        print("War mode again!!")
                        print(len(player_one.self_cards))
                        print(len(player_two.self_cards))
                        if len(player_one.self_cards) == len(player_two.self_cards):
                            print("Both have same number of cards")
                        if len(player_one.self_cards) < 4:
                            print("Player one ran out of cards")
                            break
                        if len(player_two.self_cards) < 4:
                            print("Player two ran out of cards")
                            break
                        continue


            print("Player one has %d cards now" %(len(player_one.self_cards)))
            print("Player two has %d cards now" %(len(player_two.self_cards)))
```

```python
start_game = Start_Game(player_one, player_two)
start_game.start()
```
